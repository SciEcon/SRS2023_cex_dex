## User Compare

```python
import pandas as pd
import matplotlib.pyplot as plt

# cex
df_cex = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Queried_Data/cex_to_dex_query.csv')

# dex
df_dex = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Queried_Data/dex_to_cex_query.csv')

fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(df_cex['Date'], df_cex['NUM_USERS'], label='CEX to DEX')
plt.plot(df_dex['Date'], df_dex['NUM_USERS'], label='DEX to CEX')

# Adding a vertical line at the date 2022-11-11
plt.axvline(x='2022/11/11', color='green', linestyle='--', label='2022-11-11')

plt.title('Number of Users over Time')
plt.xlabel('Date')
plt.ylabel('Number of Users')

xticks_interval = 7
xticks = df_cex['Date'][::xticks_interval]
plt.xticks(xticks)

plt.legend()
plt.show()
```
