## Volume of CEX

## Fee and Volume of CEX

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Processed_Data/cex_process.csv')
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(12, 6))

plt.semilogy(df['Date'], df['USD_VOLUMES'], label='TOTAL_VOLUMES')
plt.semilogy(df['Date'], df['FEES'], label='TOTAL_FEES')

plt.title('Total Volumes and Total Fees over Time of CEX')
plt.xlabel('Date')
plt.ylabel('Amount')

plt.legend()
plt.show()
```

### correlation coefficient

```python
import pandas as pd
from scipy.stats import pearsonr


df = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Processed_Data/cex_process.csv')

usd_volumes = df['USD_VOLUMES']
fees = df['FEES']


correlation_coefficient, p_value = pearsonr(usd_volumes, fees)


print("correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

### Granger causality

```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests

df = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Processed_Data/cex_process.csv')


data = pd.DataFrame({
    'fees': df['FEES'],
    'volumes': df['USD_VOLUMES']
})

results = grangercausalitytests(data, maxlag=1, verbose=False)

granger_results = results[1][0]

p_value = granger_results['ssr_ftest'][1]


if p_value < 0.05:
    causality = "consistent Granger causality"
else:
    causality = "no consistent Granger causality"

print("p-value:", p_value)
print("Granger causality:", causality)
```

## Fee and Transaction of CEX

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Processed_Data/cex_process.csv')
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 6))


plt.semilogy(df['Date'], df['FEES'], label='TOTAL_FEES')
plt.semilogy(df['Date'], df['TOTAL_TRANSACTIONS'], label='TOTAL_TRANSACTIONS')

plt.title('Total Fees and Total Transactions over Time of CEX')
plt.xlabel('Date')
plt.ylabel('Amount')

plt.legend()
plt.show()
```

### correlation coefficient

```python
import pandas as pd
from scipy.stats import pearsonr


df = pd.read_csv('https://raw.githubusercontent.com/SciEcon/SRS2023_cex_dex/main/data/Processed_Data/cex_process.csv')

total_transactions = df['TOTAL_TRANSACTIONS']
fees = df['FEES']


correlation_coefficient, p_value = pearsonr(total_transactions, fees)


print("correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)
```
